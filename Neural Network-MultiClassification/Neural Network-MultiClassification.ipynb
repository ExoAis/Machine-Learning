```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
import sklearn
from sklearn import metrics
from mlxtend.data import loadlocal_mnist
import matplotlib.pyplot as plt
```

```python
x_train, y_train = loadlocal_mnist (
    images_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/train-images.idx3-ubyte', 
    labels_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/train-labels.idx1-ubyte')
x_test, y_test = loadlocal_mnist(
    images_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/t10k-images.idx3-ubyte', 
    labels_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/t10k-labels.idx1-ubyte')
x_train = x_train[:5000]
y_train = y_train[:5000]
```

```python
model = Sequential ([
    Dense(units = 30, activation = 'relu'),
    Dense(units = 20, activation = 'relu'),
    Dense(units = 10, activation = 'linear')
])
```

```python
model.compile (
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    optimizer = tf.keras.optimizers.Adam(0.001)
)
```

```python
model.fit (
    x_train, y_train,
    epochs = 500
)
```

```python
prediction = model.predict(x_test)
def softmax(prediction):
    m, n = prediction.shape
    array = []
    for i in range(m * n):
        array.append(0)
    array = np.arange(100000).reshape(10000,10)
    for i in range(m):
        nume = np.exp(prediction[i])
        su = nume / (np.sum(np.exp(prediction[i])))
        array[i] = su
    return array
prediction_ = softmax(prediction)
prediction__ = tf.nn.softmax(prediction)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

m, n = x_test.shape
fig, axes = plt.subplots(8,8, figsize=(8,8))
fig.tight_layout(pad=0.1)
for i,ax in enumerate(axes.flat):
    random_index = np.random.randint(m)
    x_test_random_reshaped = x_test[random_index].reshape((28,28)).T
    ax.imshow(x_test_random_reshaped, cmap='gray')
    ax.set_title(np.argmax(prediction__[random_index]))
    ax.set_axis_off()
```

```python
prediction___ = []
for i in range(len(prediction__)):
    prediction___.append(np.argmax(prediction__[i]))
```

```python
accuracy = metrics.accuracy_score(y_test, prediction___)
print("The accuracy of the 0-9 Digit Classification Neural Network is", accuracy * 100, "%")
```
