```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
import matplotlib.pyplot as plt
import pandas as pd

```

```python
x_train = pd.read_csv('C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/SingleVariableLinearRegression/SinVarLinReg.csv', usecols = ['Temp'])
y_train = pd.read_csv('C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/SingleVariableLinearRegression/SinVarLinReg.csv', usecols = ['Salinity'])
x_test = x_train[2501:100000]
y_test = y_train[2501:100000]
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
x_test = x_test.to_numpy()
y_test = y_test.to_numpy()
x_train = x_train[:2500]
y_train = y_train[:2500]
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
model = Sequential([
    tf.keras.Input(shape = (2500, 1)),
    Dense(units = 1080, activation = 'relu'),
    Dense(units = 720, activation = 'relu'),
    Dense(units = 360, activation = 'linear')
])
```

```python
model.compile (
    loss = tf.keras.losses.MeanSquaredError(),
    optimizer = tf.keras.optimizers.Adam(0.01)
)
```

```python
model.fit (
    x_train,
    y_train,
    epochs = 10
)
```

```python
prediction = model.predict(x_test)
```

```python
def accuracy(y_pred, Y):
    for i in range(50):
        print(Y[i], " ", y_pred[i])
```

```python
print("Some values are ", accuracy(prediction, y_test))
```
