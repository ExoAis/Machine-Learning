```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
from mlxtend.data import loadlocal_mnist
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_train, y_train = loadlocal_mnist(
images_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/train-images.idx3-ubyte', 
labels_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/train-labels.idx1-ubyte')
x_test, y_test = loadlocal_mnist(
images_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/t10k-images.idx3-ubyte', 
labels_path='C:/Users/Muhammad Darab/OneDrive/Desktop/Online Course Files/Machine Learning/Data Sets/NeuralNetwork-LogisticsDigits/t10k-labels.idx1-ubyte')
x_train = x_train[:20000]
y_train = y_train[:20000]
x_test = x_test[:10000]
y_test = y_test[:10000]
```

```python
m, n = x_train.shape
for i in range(m):
    for j in range(m):
        if y_train[j] != 0 and y_train[j] != 1:
            x_train = np.delete(x_train, j, axis = 0)
            y_train = np.delete(y_train, j, axis = 0)
            m, n = x_train.shape
            break
m, n = x_test.shape
for i in range(m):
    for j in range(m):
        if y_test[j] != 0 and y_test[j] != 1:
            x_test = np.delete(x_test, j, axis = 0)
            y_test = np.delete(y_test, j, axis = 0)
            m, n = x_test.shape
            break
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
model = Sequential ([
    Dense(units = 50, activation = 'sigmoid'),
    Dense(units = 30, activation = 'sigmoid'),
    Dense(units = 1, activation = 'sigmoid')
])
```

```python
model.compile (
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(0.01)
)
```

```python
model.fit (
    x_train, y_train,
    epochs = 500
)
```

```python
y_pred = model.predict(x_test)
```

```python
for i in range(len(y_pred)):
    if y_pred[i] >= 0.5:
        y_pred[i] = 1
    else:
        y_pred[i] = 0
```

```python
def F1Score_Accuracy(Y_, Y):
    prec = 0
    recall = 0
    TP = 0
    FP = 0
    FN = 0
    TN = 0
    for i in range(len(Y)):
        if Y_[i] == Y[i] and Y_[i] == 1:
            TP += 1
        elif Y_[i] != Y[i] and Y_[i] == 1:
            FP += 1
        elif Y_[i] != Y[i] and Y_[i] == 0:
            FN += 1
        elif Y_[i] == Y[i] and Y_[i] == 0:
            TN += 1
    prec = TP / (TP + FP)
    recall = TP / (TP + FN)
    F1Score = (2 * prec * recall) / (prec + recall)
    accuracy = (TP + TN) / (TP + FN + TN + FP)
    return F1Score, accuracy
```

```python
F1Score, Accuracy = F1Score_Accuracy(y_pred, y_test)
print("F1 Score and Accuracy of the Model is", round(F1Score * 100, 2), "%", "and", round(Accuracy * 100, 2), "%", "respectively")
```
